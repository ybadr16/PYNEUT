```python
import os
import openmc.data
import numpy as np
from src.cross_section_read import CrossSectionReader
from src.vt_calc import VelocitySampler

# --- CONFIGURATION ---
# Define the path to your ENDF/B data directory
# Assuming structure is: ./endfb/neutron/Pb208.h5
BASE_DATA_PATH = "./endfb" 
H5_FILE_PATH = os.path.join(BASE_DATA_PATH, "neutron", "Pb208.h5")

def compare_xs():
    print(f"Loading OpenMC data from: {H5_FILE_PATH}")
    
    if not os.path.exists(H5_FILE_PATH):
        raise FileNotFoundError(f"Could not find the data file at {H5_FILE_PATH}. Check your directory structure.")

    # 1. Setup Mass (AMU -> kg conversion)
    # openmc.data.atomic_mass returns mass in AMU
    amu_mass = openmc.data.atomic_mass('Pb208')
    amu_to_kg = 1.66053906660e-27
    mass_kg = amu_mass * amu_to_kg
    
    # 2. Initialize Code's Components
    reader = CrossSectionReader(BASE_DATA_PATH)
    sampler = VelocitySampler(mass=mass_kg)
    
    # 3. Load OpenMC Data Directly
    # We load the HDF5 file directly to bypass the need for cross_sections.xml for this specific check
    omc_data = openmc.data.IncidentNeutron.from_hdf5(H5_FILE_PATH)
    
    # Define energies to check (eV)
    # 10 keV, 100 keV, 500 keV, 1 MeV, 2 MeV
    energies = [1e4, 1e5, 5e5, 1e6, 2e6] 
    
    print("-" * 75)
    print(f"{'Energy (eV)':<15} {'My Sigma_t':<15} {'OpenMC Sigma_t':<15} {'Diff %':<10}")
    print("-" * 75)
    
    for E in energies:
        # --- A. Get Your Cross Sections ---
        # N=1.0 ensures we are comparing microscopic cross-sections (effectively)
        _, _, _, my_sigma_t = reader.get_cross_sections(
            element="Pb208", 
            energy=E, 
            sampler=sampler, 
            number_density=1.0
        )
        
        # --- B. Get OpenMC Cross Sections ---
        # OpenMC XS are stored by MT number. 
        # MT=2 (Elastic) + MT=102 (Capture)
        # Note: OpenMC returns XS in barns. This code usually handles barns to cm2 internally 
        # or in specific functions. 
        # CHECK: My reader.get_cross_sections returns Macroscopic (cm^-1) because you pass N.
        # Let's get OpenMC result in Barns first
        omc_elastic = omc_data[2].xs['294K'](E)
        omc_capture = omc_data[102].xs['294K'](E)
        omc_micro_total_barns = omc_elastic + omc_capture
        
        # Convert OpenMC to Macroscopic (cm^-1) with N=1.0 to match our function return
        # 1 barn = 1e-24 cm^2
        omc_sigma_t = omc_micro_total_barns * 1.0 * 1e-24
        
        # --- C. Compare ---
        # Handle division by zero if XS is 0
        if omc_sigma_t > 0:
            diff = 100 * abs(my_sigma_t - omc_sigma_t) / omc_sigma_t
        else:
            diff = 0.0 if my_sigma_t == 0 else 100.0
            
        print(f"{E:<15.1e} {my_sigma_t:<15.5e} {omc_sigma_t:<15.5e} {diff:<10.2f}")

if __name__ == "__main__":
    compare_xs()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import openmc.data
from src.cross_section_read import CrossSectionReader
from src.vt_calc import VelocitySampler
import os

def debug_plot():
    # Setup
    base_path = "./endfb"
    h5_path = os.path.join(base_path, "neutron", "Pb208.h5")
    
    # Check file
    if not os.path.exists(h5_path):
        print(f"Error: {h5_path} not found.")
        return

    # Load OpenMC Data
    print("Loading OpenMC Ground Truth...")
    omc_data = openmc.data.IncidentNeutron.from_hdf5(h5_path)
    
    # Initialize My Reader
    reader = CrossSectionReader(base_path)
    pb_mass = openmc.data.atomic_mass('Pb208') * 1.66e-27
    sampler = VelocitySampler(mass=pb_mass)
    
    # Create Energy Grid (Logspace to capture low and high)
    energies = np.logspace(3, 6.5, 500) # 1 keV to ~3 MeV
    
    my_sigma_t = []
    omc_sigma_t = []
    
    print("Sampling cross sections...")
    for E in energies:
        # 1. My Code
        _, _, _, my_val = reader.get_cross_sections("Pb208", E, sampler, 1.0)
        my_sigma_t.append(my_val)
        
        # 2. OpenMC (Barns to cm^-1)
        omc_val = (omc_data[2].xs['294K'](E) + omc_data[102].xs['294K'](E)) * 1e-24
        omc_sigma_t.append(omc_val)
        
    # Plot
    plt.figure(figsize=(10, 6))
    plt.loglog(energies, omc_sigma_t, 'k-', label='OpenMC (Ground Truth)', alpha=0.8)
    plt.loglog(energies, my_sigma_t, 'r--', label='My Reader', linewidth=1.5)
    
    plt.xlabel("Energy (eV)")
    plt.ylabel("Macroscopic XS (cm^-1) [N=1.0]")
    plt.title("Diagnostics: Pb-208 Cross Section Mismatch")
    plt.grid(True, which="both", linestyle='--', alpha=0.5)
    plt.legend()
    plt.savefig("xs_debug.png")
    plt.show()

if __name__ == "__main__":
    debug_plot()
```
